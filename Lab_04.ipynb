```python
import pandas as pd
```

```python
col_data={'color':['Red','Green','Blue','Yello','Black','White']}
df_col=pd.DataFrame(col_data)
```

```python
one_hot_encoded=pd.get_dummies(df_col['color'],prefix='color')
df_col=df_col.join(one_hot_encoded)
```

```python
one_hot_encoded=one_hot_encoded.astype(int)
```

```python
print(one_hot_encoded)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
df_col['color_label']=label_encoder.fit_transform(df_col['color'])
```

```python
from datetime import datetime
```

```python
emp['DOB']=pd.to_datetime(emp['DOB'],format='%d-%m-%Y')
```

```python
emp.info()
```

```python
today = pd.Timestamp.now()
```

```python
today
```

```python
emp['Age'] = today.year - emp['DOB'].dt.year - ((today.month < emp['DOB'].dt.month) | ((today.month == emp['DOB'].dt.month) & (today.day < emp['DOB'].dt.day))).astype(int)
```
