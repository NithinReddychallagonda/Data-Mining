```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x=[1,3,5,7,9]
y=[2,4,6,8,10]
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['age'], y=df['chol'])
plt.title('Scatter Plot')
plt.show()
```

```python
data = pd.DataFrame({
    'Math': [85, 78, 92, 88, 76],
    'Science': [88, 79, 95, 84, 82],
    'English': [75, 83, 89, 92, 80]
})

correlation_matrix = data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x=df['age'], y=df['chol'])
plt.title('Line Plot: Age vs. Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df['age'], y=df['chol'])
plt.title('Bar Plot: Age Type vs. Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```
