```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
#example data
data={'category':['A','B','C','A','B','C','A','B']}
df=pd.DataFrame(data)
df
```

```python
#label encoder
label_encoder=LabelEncoder()
df['label_Encoder']=label_encoder.fit_transform(df['category'])
df
```

```python
#one_Hot Encoder
one_hot_encoder=pd.get_dummies(df['category'],prefix='category')
df=pd.concat([df,one_hot_encoder],axis=1)
print(df)
```

#Creating Deriverd Columns

```python
from datetime import datetime
```

```python
#example data
data={'Name':['Nithin','Bob'], 'Date_of_Birth':['2005-10-07','1995-08-20']}
df=pd.DataFrame(data)
df
```

```python
#converting date_of_birth to datetime
df['Date_of_Birth']=pd.to_datetime(df['Date_of_Birth'])
```

```python
#calculate age
current_year= datetime.now().year
df['Age']=current_year-df['Date_of_Birth'].dt.year
df
```

#Group by and aggregate

```python
#example data
data={'category':['A','A','B','B','C'],
      'Value':[10,20,30,40,50]}
df=pd.DataFrame(data)
df
```

```python
#group by and aggregate
grouped=df.groupby('category').agg({'Value':{'sum','mean'}})
print("Grouped data\n",grouped)
```

```python
#pivot table
pivot_table=df.pivot_table(index='category',values='Value',aggfunc=['sum','mean'])
print("\nPivot Table\n",pivot_table)
```
